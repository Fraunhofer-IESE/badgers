# Generate trends in time series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from badgers.generators.time_series.trends import GlobalAdditiveLinearTrendGenerator
```

## Setup random generator

```python
from numpy.random import default_rng
seed = 0
rng = default_rng(seed)
```

## Generate data (gaussian white noise)

```python
X = pd.DataFrame(data=rng.normal(size=(100, 2)), columns=['dimension_0', 'dimension_1'])
```

```python
X.plot(subplots=True)
```

## Add linear trend

```python
generator = GlobalAdditiveLinearTrendGenerator(random_generator=rng)
```

```python
Xt, _ = generator.generate(X, y=None, slope=np.array([4,-2]))
```

```python
fig, axes = plt.subplots(2, sharex=True, sharey=True, figsize=(6,6))
X.plot(ax=axes[0])
axes[0].set_title('Original data')
Xt.plot(ax=axes[1])
axes[1].set_title('Transformed data')
plt.tight_layout();
```
